# ***Data Cleaning***

```python
df['HS'] = df['HS'].str.strip('*')
print(df)
```

```python
df.rename(columns={'Player Name':'Player'}, inplace=True)
print(df)
```

```python
df
```

```python
df=df.drop(columns=['100s','No']) # These columns were already dropped
print(df)
```

```python
df['Team'] = df['Team'].str.strip()
print(df)
```

```python
df
```

```python
df['HS'] = df['HS'].fillna(2)
print(df)
```

```python
df.duplicated().sum()
print(df)
```

```python
df['BF'] = df['BF'].astype(float)
print(df)
```

```python
df.reset_index(drop=True, inplace=True)
print(df)
```

```python
df.to_excel("IPL2025_Cleaned.xlsx", index=False)
```

# ***EDA***

```python
plt.figure(figsize=(10,5))
plt.bar(df['Player'][:10], df['Runs'][:10], color='orange')
plt.xticks(rotation=90)
plt.xlabel("Player")
plt.ylabel("Runs")
plt.title("Top 10 Run Scorers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Runs'], bins=10, color='green', edgecolor='black')
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.title("Distribution of Runs")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Matches'], df['Runs'], color='red', marker='o')
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.title("Runs vs Matches")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['SR'], df['Runs'])
plt.xlabel("Strike Rate")
plt.ylabel("Runs")
plt.title("SR vs Runs")
plt.show()
```

```python
team_runs = df.groupby('Team')['Runs'].sum()

plt.figure(figsize=(7,7))
# Define colors for all teams (assuming 10 teams based on team_runs)
colors = ['gold','skyblue','lightgreen','pink','violet','orange','cyan','magenta','brown','teal']
# Create an explode tuple with the same length as the number of teams
explode = (0.1,) + (0,) * (len(team_runs) - 1)

plt.pie(team_runs, explode=explode, labels=team_runs.index, autopct='%1.1f%%',
        colors=colors, shadow=True, startangle=140)
plt.title("Team Wise Runs")
plt.axis('equal')
plt.show()
```
