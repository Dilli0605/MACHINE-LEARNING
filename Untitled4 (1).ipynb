```python
df.drop(columns=["Remarks","Temp_Code"])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(df)
```

```python
df['City'] = df['City'].fillna("Unknown")
print(df)
```

```python
df['Purchase_Amount'] = df['Purchase_Amount'].fillna(df['Purchase_Amount'].median())
print(df)
```

```python
df = df.drop(columns=['Temp_Code','Remarks'], errors='ignore')
print(df)
```

```python
df = df.drop_duplicates()
print(df)
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df = df.rename(columns={
    'Phone_Number': 'Phone',
    'Purchase_Amount': 'Amount'
})
print(df)
```

```python
df['Name'] = df['Name'].str.strip()
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')
print(df)
```

```python
df['Email'] = df['Email'].fillna("Unknown")
print(df)
```

```python
df['Join_Date'] = df['Join_Date'].ffill()
print(df)
```

```python
df['Phone'] = df['Phone'].astype(str)
df['Phone'] = df['Phone'].str.replace(r'\.0$', '', regex=True)
df['Phone'] = df['Phone'].str.replace(r'\D', '', regex=True)
df['Phone'] = df['Phone'].apply(lambda x: x if len(x)==10 else np.nan)
df['Phone'] = df['Phone'].fillna('-')
print(df)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_excel("cleaned_customer_data.xlsx", index=False)
```

# ***MatplotgLib*** ***Plot***

```python
#Line Plot
plt.figure(figsize=(8,5))
plt.plot(df['Customer_ID'], df['Amount'])
plt.xlabel("Customer ID")
plt.ylabel("Purchase Amount")
plt.title("Purchase Trend")
plt.show()
```

```python
#Bar Plot
city_counts = df['City'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(city_counts.index, city_counts.values, color='orange')
plt.xlabel("City")
plt.ylabel("Count")
plt.title("Customers per City")
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram Plot
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10, color='purple', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Amount'], color='red')
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.title("Age vs Purchase Amount")
plt.show()
```

```python
#Pie Chart
gender_count = df['Gender'].value_counts()
colors = ['gold','lightcoral','lightskyblue']
plt.figure(figsize=(6,6))
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%', colors=colors)
plt.title("Gender Distribution")
plt.axis('equal')
plt.show()
```

# ***Seaborn Plot***

```python
#Count Plot
plt.figure(figsize=(8,5))
sns.countplot(x='City', data=df, palette='Blues')
plt.title("City Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram Plot
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True, color='green')
plt.title("Age Distribution")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Amount', hue='Gender', data=df, palette='cool')
plt.title("Age vs Amount")
plt.show()
```

```python
#Heatmap plot
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='viridis')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(x='Gender', y='Amount', data=df, palette='Set2')
plt.title("Amount by Gender")
plt.show()
```
