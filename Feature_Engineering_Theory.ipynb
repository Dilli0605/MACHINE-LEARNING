## Categorical Encoding

### One-Hot Encoding

One-hot encoding creates a new binary column for each unique category in a categorical feature.
Each column contains 1s and 0s, indicating the presence of a category.

Suitable for nominal (unordered) categorical features where no ordinal relationship exists between the categories (e.g., colors, city names, etc.).
